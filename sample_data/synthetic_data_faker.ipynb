```python
import csv
import random
from faker import Faker
import pandas as pd
```

```python
fake = Faker()
num_rows = 50

# --- Generate synthetic data ---
rows = []
for _ in range(num_rows):
    row = {}
    for col, sensitivity in columns:
        if sensitivity == "Restricted":
            if "ssn" in col:
                row[col] = fake.ssn()
            elif "passport" in col:
                row[col] = fake.passport_number()
            elif "dob_full" in col:
                row[col] = fake.date_of_birth().strftime("%Y-%m-%d")
        elif sensitivity == "Confidential":
            if "dob_year_month" in col:
                row[col] = fake.date_of_birth().strftime("%Y-%m")
            elif "budget" in col:
                row[col] = f"${random.randint(1000,10000)}"
        elif sensitivity == "Internal-use only":
            row[col] = fake.sentence(nb_words=6)
        elif sensitivity == "Public":
            if "job_title" in col:
                row[col] = fake.job()
            elif "press_release" in col:
                row[col] = fake.sentence(nb_words=10)
    rows.append(row)

# --- Write to CSV ---
output_path = '/Users/clu/Library/CloudStorage/OneDrive-AldoGroupInc/Desktop/Metascout/agentic models/synthetic data/synthetic_user_records.csv'
with open(output_path, "w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=[col[0] for col in columns])
    writer.writeheader()
    writer.writerows(rows)

print(f"Synthetic dataset saved to {output_path}")

```
